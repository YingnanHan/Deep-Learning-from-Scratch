```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#生成数据
x = np.arange(0,6,0.1) # 以0.1为单位 生成0-6的数据
y = np.sin(x)

#绘制图形
plt.plot(x,y)
plt.show()
```

#### pyplot的功能

```python
#生成数据
x = np.arange(0,6,0.1) # 以0.1为单位 生成0-6的数据
y1 = np.sin(x)
y2 = np.cos(x)

#绘制图形
plt.plot(x,y1,label="sin",linestyle="solid")  #用实线绘制
plt.plot(x,y2,label="cos",linestyle="dashed") #用虚线绘制
plt.xlabel("x")#x轴标签
plt.ylabel("y")#y轴标签
plt.title("sin & cos")#标题
plt.legend()
plt.show()
```

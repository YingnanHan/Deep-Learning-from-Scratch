#### 梯度求解函数

```python
def numerical_gradient(f,x):
    h = 1e-4 # 0.0001
    grad = np.zeros_like(x) # 生成和x形状相同的 值为0的数组
    
    for idx in range(x.size):
        tmp_val = x[idx]
        # f(x+h)的计算
        x[idx] = tmp_val + h
        fxh1 = f(x)
        
        #f(x-h)的计算
        x[idx]  = tmp_val - h
        fxh2 = f(x)
        
        grad[idx] = (fxh1-fxh2)/(2*h) # 保存关于某一个自变量的偏导数
        x[idx] = tmp_val #还原值
    return grad
```

```python
#实现 y = x0^2 + x1^2
def function_2(x):
     return x[0]**2 + x[1]**2
#或者 return np.sum(x**2)
```

```python
numerical_gradient(function_2,np.array([3.0,4.0]))
```

##### 梯度图像向量场二维图像的绘制

```python
# coding: utf-8
# cf.http://d.hatena.ne.jp/white_wheels/20100327/p3
import numpy as np
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D


def _numerical_gradient_no_batch(f, x):
    h = 1e-4 # 0.0001
    grad = np.zeros_like(x)
    
    for idx in range(x.size):
        tmp_val = x[idx]
        x[idx] = float(tmp_val) + h
        fxh1 = f(x) # f(x+h)
        
        x[idx] = tmp_val - h 
        fxh2 = f(x) # f(x-h)
        grad[idx] = (fxh1 - fxh2) / (2*h)
        
        x[idx] = tmp_val # 还原值
        
    return grad


def numerical_gradient(f, X):
    if X.ndim == 1:
        return _numerical_gradient_no_batch(f, X)
    else:
        grad = np.zeros_like(X)
        
        for idx, x in enumerate(X):
            grad[idx] = _numerical_gradient_no_batch(f, x)
        
        return grad


def function_2(x):
    if x.ndim == 1:
        return np.sum(x**2)
    else:
        return np.sum(x**2, axis=1)


def tangent_line(f, x):
    d = numerical_gradient(f, x)
    print(d)
    y = f(x) - d*x
    return lambda t: d*t + y
     
if __name__ == '__main__':
    x0 = np.arange(-2, 2.5, 0.25)
    x1 = np.arange(-2, 2.5, 0.25)
    X, Y = np.meshgrid(x0, x1)
    
    X = X.flatten()
    Y = Y.flatten()
    
    grad = numerical_gradient(function_2, np.array([X, Y]) )
    
    plt.figure()
    plt.quiver(X, Y, -grad[0], -grad[1],  angles="xy",color="#666666")#,headwidth=10,scale=40,color="#444444")
    plt.xlim([-2, 2])
    plt.ylim([-2, 2])
    plt.xlabel('x0')
    plt.ylabel('x1')
    plt.grid()
    plt.legend()
    plt.draw()
    plt.show()
```

###### 在上式中η表示更新量，在神经网络的学习中被称为学习率，决定了在一次学习中应该学习多少，以及多大程度上更新参数

###### 值得注意的是尽管上面描述的仅仅是一个变量的更新表达式，但是这个式子对于任何变量都是起作用的

#### 使用梯度下降法

```python
def gradient_descent(f,init_x,lr=0.01,step_num=100):
    #lr:表示学习率 init_x是初始值 step_num是重复次数
    x = init_x
    
    for i in range(step_num):
        grad = numerical_gradient(f,x)
        x -= lr*grad #更新x，便于下一次进行梯度的计算
    return x  # 理论上x grad均可以是多维向量  
```

#### 实例：求解指定函数f(x) = x₁²+x₂²的最小值

```python
def function_2(x):
    return x[0]**2 + x[1]**2
```

```python
init_x = np.array([-3.0,4.0]) #设定从二维平面点 -3,4开始进行梯度下降运算
gradient_descent(function_2,init_x,0.1,step_num=100)#如何确定 step以及step_num？
```

#### 绘制函数梯度求解函数的平面点更新过程

```python
# coding: utf-8
import numpy as np
import matplotlib.pylab as plt


def gradient_descent(f, init_x, lr=0.01, step_num=100):
    x = init_x
    x_history = []

    for i in range(step_num):
        x_history.append( x.copy() )

        grad = numerical_gradient(f, x)
        x -= lr * grad

    return x, np.array(x_history)


def function_2(x):
    return x[0]**2 + x[1]**2

init_x = np.array([-3.0, 4.0])    

lr = 0.1
step_num = 20
x, x_history = gradient_descent(function_2, init_x, lr=lr, step_num=step_num)

plt.plot( [-5, 5], [0,0], '--b')
plt.plot( [0,0], [-5, 5], '--b')
plt.plot(x_history[:,0], x_history[:,1], 'o')

plt.xlim(-3.5, 3.5)
plt.ylim(-4.5, 4.5)
plt.xlabel("X0")
plt.ylabel("X1")
plt.show()
```

#### 学习率出现偏差的例子以及导致的结果

```python
def gradient_descent(f,init_x,lr=0.01,step_num=100):
    #lr:表示学习率 init_x是初始值 step_num是重复次数
    x = init_x
    
    for i in range(step_num):
        grad = numerical_gradient(f,x)
        x -= lr*grad #更新x，便于下一次进行梯度的计算
    return x  # 理论上x grad均可以是多维向量
```

```python
#学习率过大的例子 lr = 10
init_x = np.array([-3.0,4.0])
gradient_descent(function_2,init_x=init_x,lr=10.0,step_num=100)
```

```python
#学习率过小的例子 lr = 1e-10
init_x = np.array([-3.0,4.0])
gradient_descent(function_2,init_x=init_x,lr=1e-10,step_num=100)
```
